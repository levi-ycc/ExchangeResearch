```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('fitx.txt')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.strftime('%H:%M:%S')
```

```python
df
```

```python
df[(df['Date']=='2019-12-12') & (df['Time']<'06:44:00')] #05:00:00 #13:45:00
```

```python
df[(df['Time']>='08:46:00') & (df['Time']<='13:45:00')] #08:46:00 ~ 13:45:00, 15:01:00~05:00:00
```

```python
df_before = df[(df['Time']>='08:46:00') & (df['Time']<='13:45:00')].copy()
df_before.index = range(df_before.shape[0])
min_2_groups = np.array_split(df_before, df_before.shape[0]/2)
```

```python
to_1_group = pd.DataFrame([[df['Date'].iloc[-1], df['Time'].iloc[-1],
                            df['Open'].iloc[0], df['High'].max(), df['Low'].min(),
                            df['Close'].iloc[-1], df['TotalVolume'].sum()] for df in min_2_groups],
                          columns=df.columns)
```

```python
to_1_group.to_csv('fitx_before_2min.csv', index=False)
```

```python
df_after = df[~((df['Time']>='08:46:00') & (df['Time']<='13:45:00'))].copy()
df_after.index = range(df_after.shape[0])
min_2_groups_after = np.array_split(df_after, df_after.shape[0]/2)
```

```python
to_1_group_after = pd.DataFrame([[df['Date'].iloc[-1], df['Time'].iloc[-1],
                            df['Open'].iloc[0], df['High'].max(), df['Low'].min(),
                            df['Close'].iloc[-1], df['TotalVolume'].sum()] for df in min_2_groups_after],
                          columns=df.columns)
to_1_group_after.to_csv('fitx_after_2min.csv', index=False)
```

```python
all_2min = pd.concat([to_1_group, to_1_group_after], axis=0)
```

```python
all_2min.sort_values(['Date', 'Time']).to_csv('fitx_2min.csv', index=False)
```
