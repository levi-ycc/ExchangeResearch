```python
from lib.utils import FeatureEngineer
import numpy as np
import pandas as pd
import talib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/txf1.csv')
```

```python
rsi15 = talib.RSI(df['C'], 15).dropna()
rsi10 = talib.RSI(df['C'], 10).dropna()[-rsi15.shape[0]:]
rsi5 = talib.RSI(df['C'], 5).dropna()[-rsi15.shape[0]:]
```

```python
plt.hist(rsi5/rsi10, bins=100)
```

```python
plt.hist((df['C']/df['C'].shift(1)).dropna(), bins=100)
```

```python
EMA15=talib.EMA(df['C'],15)
EMA10=talib.EMA(df['C'],10)[-EMA15.shape[0]:]
EMA5=talib.EMA(df['C'],5)[-EMA15.shape[0]:]
```

```python
plt.hist(EMA5/EMA15, bins=100, range=[0.985, 1.010])
```
